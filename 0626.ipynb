* 정해진 횟수만큼 반복

**print()함수 옵션**; 여러항목 출력하기

**range()함수**  
range(n); 0부터 n-1까지, n개의 정수 생성.  
range(m,n); m부터 n-1까지, n-m개의 정수 생성.  
range(m,n,p); m부터 n-1까지 p의 간격을 가지고 정수 생성.  
**round(n, m)**  
소숫점 m째 자리에서 반올림.

```python
# 가위바위보 ㅍ로그램
import random
while 1:
    while 1:
        user_input = (input("가위, 바위, 보? "))
        if (user_input=='가위' or user_input=='바위' or user_input=='보'):
            break
        print("[입력오류] 가위, 바위, 보 중 선택하세요.")

    cpt_random = random.randint(0,2)
    if cpt_random==0:
        cpt_input='가위'
    elif cpt_random==1:
        cpt_input='바위'
    else:
        cpt_input='보'

    print("user:{0}, computer:{1}".format(user_input, cpt_input))

    if user_input==cpt_input:
        print("비겼당..")
    elif user_input=='가위':
        if cpt_input=='보':
            print("이겼당~~")
        else:
            print("졌당............")
    elif user_input=='바위':
        if cpt_input=='가위':
            print("이겼당~~")
        else:
            print("졌당............")
    elif user_input=='보':
        if cpt_input=='바위':
            print("이겼당~~")
        else:
            print("졌당............")
    
    conti_ = input("계속할래?(ㄴㄴ입력시 그만) ")
    if conti_=='ㄴㄴ':
        break
    print("=====restart=====")
```

```python
# HW1 주민등록번호를 숫자로 입력하면 해당 주민등록 번호에서 지역번호 부분을 이용하여 
# 출생지를 출력하는 프로그램

p_num = input("주민등록번호를 (-)를 빼고 입력하세요: ")
r_num = (int(p_num)%1000000)//10000

if r_num>95:
    print("잘못된 주민등록번호입니다.")
else:
    if r_num>=00 and r_num<=8:
        hometown = '서울'
    elif r_num<=12:
        hometown = '부산'
    elif r_num<=15:
        hometown = '인천'
    elif r_num<=25:
        hometown = '경기'
    elif r_num<=34:
        hometown = '강원'
    elif r_num<=47:
        hometown = '충청'
    elif r_num<=66:
        hometown = '전라'
    elif r_num<=91:
        hometown = '경상'
    elif r_num<=95:
        hometown = '제주'
    print(p_num,": 출생 지역은",hometown,"입니다.")
```

```python
# HW2 주사위를 던져서 각 면이 나오는 횟수 및 확률을 구하는 프로그램

#화면상에서 주사위를 던질 회수를 입력하면 그 회수만큼 for 문장을 이용하여 반복하면서 랜덤하게 1에서 6까지의 숫자 생성한다.
#최소 100회 이상 주사위를 던질 숫자를 입력하고, 100회 보다 작은 수를 입력하는 경우 프로그램을 종료함
#주사위를 던지는 횟수가 많아질 수록 각 면이 나오는 확률이 1/6 (약 0.167)이 되는지확인 하시오. (확률 계산은 소수점 3자리까지 계산)

import random

count = int(input("주사위를 던질 회수를 입력하세요 (100 이상): "))
if count>=100:
    
    face_1 = 0
    face_2 = 0
    face_3 = 0
    face_4 = 0
    face_5 = 0
    face_6 = 0
    
    for i in range(count):
        face = random.randint(1,6)
        if face == 1:
            face_1 += 1
        elif face == 2 :
            face_2 += 1
        elif face == 3 :
            face_3 += 1
        elif face == 4 :
            face_4 += 1
        elif face == 5 :
            face_5 += 1
        else:
            face_6 += 1

    print("주사위면 1: {0}회/{1}, 확률: {2}".format(face_1, count, round(face_1/count, 3)))
    print("주사위면 2: {0}회/{1}, 확률: {2}".format(face_2, count, round(face_2/count, 3)))
    print("주사위면 3: {0}회/{1}, 확률: {2}".format(face_3, count, round(face_3/count, 3)))
    print("주사위면 4: {0}회/{1}, 확률: {2}".format(face_4, count, round(face_4/count, 3)))
    print("주사위면 5: {0}회/{1}, 확률: {2}".format(face_5, count, round(face_5/count, 3)))
    print("주사위면 6: {0}회/{1}, 확률: {2}".format(face_6, count, round(face_6/count, 3)))
    
else:
    print("100 이상의 숫자를 입력하세요. 종료합니다.")
```
