```python
# 고양이 클래스(private변수와 getter,setter)
class Cat:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
    def setName(self, name):
        self.__name = name
    def getName(self):
        return self.__name
    def setAge(self, age):
        self.__age = age
    def getAge(self):
        return self.__age

deung_a = Cat('등어', 2)
print(deung_a.getName(), deung_a.getAge())
```

```python
# 특수메소드
class Vector2D :
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Vector2D(self.x + other.x, self.y + other.y)
    def __sub__(self, other):
        return Vector2D(self.x - other.x, self.y - other.y)
    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    def __str__(self):
        return '(%g, %g)' % (self.x, self.y)
u = Vector2D(0,1)
v = Vector2D(1,0)
w = Vector2D(1,1)
a = u + v
print(a)
```

```python
# HW7 주민등록번호 정상판별 프로그램

class IDnumber:
    def __init__(self, n):
        self.n = n
        
    def check(self):
        s = 0
        for i in range(12):
            s += int(self.n[i])*(i%8+2)
        if int(self.n[-1])==(11-(s%11))%10:
            return True
        else:
            return False
    
    def adult_check(self):
        d = 2019-(1900+self.n[0]*10+self.n[1])
        if d>=19:
            return '성인 인증이 되었습니다. 나이 %d세' % (d)
        else:
            return '성인 인증이 실패하였습니다.'

    def getBday(self):
        b_day = int(self.n)//10000000
        return b_day

    def getHome(self):
        r_num = (int(self.n)%1000000)//10000
        if r_num>=00 and r_num<=8:
            hometown = '서울'
        elif r_num<=12:
            hometown = '부산'
        elif r_num<=15:
            hometown = '인천'
        elif r_num<=25:
            hometown = '경기'
        elif r_num<=34:
            hometown = '강원'
        elif r_num<=47:
            hometown = '충청'
        elif r_num<=66:
            hometown = '전라'
        elif r_num<=91:
            hometown = '경상'
        elif r_num<=95:
            hometown = '제주'
        return hometown
    
def main():
    while 1:
        n = input("주민등록번호를 숫자만 입력하세요: ")
        if len(n)>13:
            print("주민등록번호가 13자리 보다 큽니다. 다시 입력하세요.")
        elif len(n)<13:
            print("입력된 숫자의 자리수가 13자리 이하입니다. 다시 입력하세요.")
        else:
            cnt = [0,0,0]
            for i in n:
                if i == '-':
                    cnt[0] +=1
                elif i == ' ':
                    cnt[1] +=1
                elif i.isalpha()==True:
                    cnt[2] += 1
            if cnt[0]!=0:
                print("-를 빼고 다시 입력하세요.")
            elif cnt[1]!=0:
                print("공백 문자가 포함되었습니다. 다시 입력하세요.")
            elif cnt[2]!=0:
                print("알파벳 문자가 포함되었습니다. 다시 입력하세요.")
            else:
                p1 = IDnumber(n)
                break
            
    if p1.check()==False:
        print("불법 생성된 주민등록번호입니다.")
    else:
        print("정상적인 주민등록번호입니다.")
        print(p1.adult_check())
        print("주민등록번호:", p1.n)
        print("생년월일:", p1.getBday())
        print("출생지:", p1.getHome())
    
main()
```
