```python
# 키를 이용한 정렬
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __repr__(self):
        return "<이름: %s, 나이: %s>" % (self.name, self.age)
    def __str__(self):
        return '[이름: %s, 나이: %s]' % (self.name, self.age)
    # str은 굳이 쓸 필요 x
def keyAge(person):
    return person.age

p1 = Person("김등어", 2)
print(p1)

people = [Person("고등어", 3), Person("김등어", 2), Person("이야옹", 1)]
print(sorted(people, key=keyAge))
```

```python
# iterable 객체 예제
class PowIterator:
    def __init__(self, max=0):
        self.max = max
        self.n = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.n <= self.max:
            result = 2 ** self.n
            print("2 ** {0} = ".format(self.n), end=' ')
            self.n += 1
            return result
        else:
            raise StopIteration
a = PowIterator(5)
for i in a:
    print(i)
```

```python
# 연산자 오버로딩 예제; 책 페이지 합치기
class Book:
    def __init__(self, name='', page=0):
        self.name = name
        self.page = page
    def __str__(self):
        return self.name
    def __add__(self, other):
        return self.page + other.page
book1 = Book("등어야 야옹해", 500)
book2 = Book("귀여운 등어", 1000)
print(book1+book2)
```

```python
# 상속; 은행계좌
class BankAccount:
    def __init__(self, name, number, balance):
        self.balance = balance
        self.name = name
        self.number = number
    def withdraw(self, amount):
        self.balance -= amount
        return self.balance
    def deposit(self, amount):
        self.balance += amount
        return self.balance
    
class SavingAccount(BankAccount):
    def __init__(self, name, number, balance, interest_rate):
        super().__init__(name, number, balance)
        self.interest_rate = interest_rate
    def set_interest_rate(self, interest_rate):
        self.interest_rate = interest_rate
    def get_interest_rate(self):
        return self.interest_rate
    def add_interest(self):
        self.balance += self.balance*self.interest_rate

class CheckingAccount(BankAccount):
    def __init__(self, name, number, balance):
        super().__init__(name, number, balance)
        self.withdraw_charge = 10000
    def withdraw(self, amount):
        return BankAccount.withdraw(self, amount+ self.withdraw_charge)

a1 = SavingAccount("김등어", 123456, 10000, 0.05)
a1.add_interest()
print("저축예금의 잔액=", a1.balance)
a2 = CheckingAccount("이야옹", 123457, 2000000)
a2.withdraw(100000)
print("당좌예금의 잔액=", a2.balance)
```

```python
# HW8 list클래스에서 상속받는 RandomIntList 클래스 구현

import random
class RandomIntList(list):

    def __init__(self, start, end):
        self.start = start
        self.end = end
        self.mylist = []
        for i in range(end-start+1):
            self.mylist.append(random.randint(start, end))

    def __repr__(self):
        strindex = ''
        strmylist = '['
        for i in range(len(self.mylist)):
            strindex += '[%2d]' % i
            if i == len(self.mylist)-1:
                strmylist += '%2d ]' % self.mylist[i]
            else:
                strmylist += '%2d, ' % self.mylist[i]
        return '%s\n%s' % (strindex, strmylist)

    def replace(self, old, new):
        cnt = 0
        for i in range(len(self.mylist)):
            if self.mylist[i]==old:
                cnt += 1
                self.mylist[i] = new
        if cnt==0:
            print("%d가 리스트에 없습니다." % old)
        else:
            print("%d개의 %d가 %d로 교체되었습니다." % (cnt, old, new))

    def dremove(self):
        new_list = []
        for i in range(len(self.mylist)-1, 0, -1):
            for j in range(i-1, -1, -1):
                if self.mylist[i] not in new_list:
                    new_list.append(self.mylist[i])
        new_list.reverse()
        self.mylist = list(new_list)

    def removeall(self, x):
        cnt = 0
        for y in self.mylist:
            if y==x:
                cnt += 1
        for i in range(cnt):
            self.mylist.remove(x)
        print("%d를 %d개 삭제하였습니다." % (x, cnt))

def main():
    start, end = input("RandomIntList 생성(시작과 끝을 입력하세요): ").split()
    start = int(start)
    end = int(end)
    mylist = RandomIntList(start, end)
    print(mylist)

    while 1:
        print("RandomIntList 테스트")
        print("1. 리스트 출력")
        print("2. replace(old, new)")
        print("3. dremove(): 중복 item 제거")
        print("4. removeall(x): 모든 x 삭제")
        print("5. 종료")

        menu = int(input("메뉴를 선택하세요: "))
        if menu==1:
            print(mylist)
        elif menu==2:
            old, new = input("교체할 숫자 (old, new) 입력(공백 구분): ").split()
            old = int(old)
            new = int(new)
            mylist.replace(old, new)
            print(mylist)
        elif menu==3:
            mylist.dremove()
            print(mylist)
        elif menu==4:
            x = int(input("삭제할 숫자를 입력하세요: "))
            mylist.removeall(x)
            print(mylist)
        elif menu==5:
            break
        else:
            print("[입력오류] 1~5의 숫자를 입력하세요.")

main()
```
