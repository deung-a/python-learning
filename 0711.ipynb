### 예외처리
오류가 발생했을 때 오류를 사용자에게 알려주고 모든 데이터를 저장하게 한 후에 사용자가 깔끔하게 프로그램을 종료할 수 있게 함.

    try:
        예외가 발생할 수 있는 문장
    except 예외1:
        예외1이 발생하면 실행하는 문장
    except 예외2:
        예외2가 발생하면 실행하는 문장
    else:
        예외가 없는 경우에 실행하는 문장
    finally:
        예외발생여부와 관계없이 항상 실행되는 문장
        
* 예외 종류  
IOError; 파일을 열 수 없을때  
ImportError; 파이썬이 모듈을 찾을 수 없을때  
ValueError; 연산이나 내장함수에서 인수가 적절치 않은 값을 가질때  
KeyboardInterrupt; 사용자가 인터럽트키를 누를때  
EOFError; 내장함수가 파일의 끝을 만남 


### CSV파일을 이용한 데이터 분석

* csv파일에서 데이터 읽어오기
    1. import csv
    2. 파일 = open('파일이름', 'r'); 데이터를 문자열 형태로 읽어옴
    3. 데이터 = csv.reader(파일, delimiter=','); delimiter:구분자 설정
    4. 데이터 출력
    5. 파일.close()  
    

* 데이터 헤더 저장  
header = next(data); next() 사용

```python
# 파일 읽고 쓰기
infilename = input("입력 파일 이름: ")
outfilename = input("출력 파일 이름: ")

infile = open(infilename, "r",  encoding='UTF8')
outfile = open(outfilename, "w")

sum = 0
count = 0

for line in infile:
    dailySale = int(line)
    sum += dailySale
    count +=1

print("총매출 = "+ str(sum))
print("평균 일매출 = "+ str(sum/count))

outfile.write("총매출 = "+ str(sum)+ '\n')
outfile.write("평균 일매출 = "+ str(sum/count))

infile.close()
outfile.close()
```

```python
# 파일의 스페이스와 탭의 개수 세는 프로그램
def parse_file(path):
    infile = open(path)
    spaces = 0
    tabs = 0
    for line in infile:
        spaces += line.count(' ')#!!!!!
        tabs += line.count('\t')
    infile.close()
    return spaces, tabs
filename = input("파일 이름을 입력하시오: ")
spaces, tabs = parse_file(filename)
print("스페이스 수 = %d, 탭의 수 = %d" % (spaces, tabs))
```

```python
import csv # CSV 모듈을 불러옴
f = open('daegu.csv', 'r')
data = csv.reader(f, delimiter = ',') # 리스트형태로 저장, delimiter:구분자 설정
header = next(data) # 데이터 헤더 저장
print(header)
i = 1
for row in data:
    print(row)
    i += 1
    if(i>=5):
        break
f.close()
```

```python
# 데이터분석 예시; 최고기온 구하기
import csv

f = open('daegu.csv','r')
data = csv.reader(f)
header = next(data)
max_temp = -999 # 최고기온을 저장할 변수 초기화
max_date = '' # 최고기온의 날짜를 저장할 변수 초기화

for row in data:
    if row[-1] == '':  # 데이터가 없는 경우
        row[-1] = -999 # 가장 작은값을 입력함
    row[-1] = float(row[-1])
    
    if (max_temp < row[-1]):
        max_temp = row[-1]
        max_date = row[0]
f.close()

print("기상관측 이래 대구 최고기온이 가장 높았던 날은", max_date,"로", max_temp,"도 였습니다.")
```
